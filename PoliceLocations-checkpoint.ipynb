```python
import flask
import folium
import pandas as pd
from pandas import read_csv
from folium.plugins import MarkerCluster
from folium import plugins
from folium.plugins import HeatMap
from folium.plugins import MeasureControl
```

# PLOTTING POLICE LOCATION ON A MAP

```python
policedata= read_csv("DATA.csv")
```

```python
policedata.head()
```

```python

#add the map
m = folium.Map(
    location=[policedata.Latitude.mean(), policedata.Longitude.mean()],
    tiles='openstreetmap',zoom_start=12)
#addin clusters on the map

marker_cluster = MarkerCluster(
    name='Police Locations',
    overlay=True,
    control=False,
    icon_create_function=None
)
# Adds measurement tool to the top right of map
m.add_child(MeasureControl())

#add fullscreen plugin to the map
plugins.Fullscreen(
    position='topright',
    title='Expand me',
    title_cancel='Exit me',
    force_separate_button=True).add_to(m)

#iterate through the train data columns, category ,weekday,and id to retrive the values and pin them to the
#coordinate values markers on the map as a pop up text when  clicked

for k, row in policedata.iterrows():
    popup_text = "police station:{}"
    popup_text = popup_text.format(row["Police_Station"])
    location = policedata.Latitude.values[k], policedata.Longitude.values[k]
    marker = folium.Marker(location=location,icon=folium.Icon(color='yellow'))
    
    popup = popup_text
    folium.Popup(popup).add_to(marker)
    marker_cluster.add_child(marker)
    
   #interactive marker
    m.add_child(folium.ClickForMarker(popup="temporary marker"))
    
    # pop up location coordinates of click on the map
    m.add_child(folium.LatLngPopup())
    
#adding markers to map
marker_cluster.add_to(m)

folium.LayerControl().add_to(m)


#save map to html file
m.save("policestation.html")
m
```
