# Potting crimes against number of their reports

```python
%matplotlib inline
import flask
import folium
import pandas as pd
from pandas import read_csv
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
crimes= read_csv("Traindata.csv")
```

```python
#function to plot a graph of crime frequencies in nairobi
def plot_data(data, Crime_Type):
    l = data.groupby(Crime_Type).size().sort_values()
    fig = plt.figure(figsize=(15,8))
    plt.style.use('ggplot')
    plt.yticks(fontsize=8)
    l.plot(kind='barh',fontsize=14,color='g')
    plt.xlabel('Number of Reports')
    plt.ylabel('Type of Crimes',fontsize=10)
    plt.title('Visual Representation of Crime Reports as Reported')

plot_data(crimes,'Crime_Type')
plt.savefig('C:/Users/Andrew/.spyder2-py3/static/crime.jpg')
plt.show()
```

# Plotting Frequency of crime per location

```python
df= read_csv("Frequency.csv")
```

```python
df.head()
```

```python
# plotting location frequencies
df.location.value_counts().plot.barh(stacked=True,figsize=(30,15))

plt.xlabel('crime frequency')
plt.ylabel('location')
plt.title('relative frequency of crime per location')

plt.savefig('C:/Users/Andrew/.spyder2-py3/static/frequency.jpg')
plt.show()
```

# Plotting Crime type frequency per location

```python
#plotting location per crime category
fig = plt.figure(figsize=(17,9))
ax = sns.heatmap(pd.crosstab(df['Crime_Type'],df['location']),linewidths=.5)
plt.title('Incident Crime Type per location',fontsize=16)
plt.legend(fontsize=14)
plt.xticks(fontsize=12,rotation=60,ha='right')
plt.yticks(fontsize=12)
plt.xlabel("location")
plt.ylabel("crime Type")
plt.savefig('C:/Users/Andrew/.spyder2-py3/static/crimeloc.jpg')
plt.show()
```
